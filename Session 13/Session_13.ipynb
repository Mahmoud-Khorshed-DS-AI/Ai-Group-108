## Visualization 8: Stacked Bar Chart - Sales by Customer Type and Product Category:

 Visualizing the sales breakdown by customer type and product category using a stacked bar chart.

```python
df.sample(2)
```

```python
customer_category_sales = df.pivot_table(index='Segment',  # categorical
                                        columns='Category',
                                        values='Sales',
                                        aggfunc='sum'
                                        )
customer_category_sales
```

```python
customer_category_sales.plot(kind='bar', figsize=(10, 6)) 

plt.xlabel('Customer Type') 
plt.ylabel('Total Sales') 
plt.title('Sales by Customer Type and Product Category') 
plt.legend(title='Product Category', bbox_to_anchor=(1, 1)) 
plt.show()
```
