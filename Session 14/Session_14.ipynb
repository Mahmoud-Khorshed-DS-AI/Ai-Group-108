## Visualization 13: Line Plot - Sales by Day of the Week 

This line plot shows the total sales trend based on the day of the week. It helps analyze whether certain days of the week have higher or lower sales.

**what is the max and min Day of sales From Image ?**  

```python
df['Purchase_Date'] = pd.to_datetime(df['Purchase_Date'])
df['Day_of_Week'] = df['Purchase_Date'].dt.day_name() 

day_of_week_sales = df.groupby('Day_of_Week')['Total_Sales'].sum() 

day_of_week_sales.plot(kind='line',
                        marker='o', figsize=(5, 3)) 

plt.xlabel('Day of the Week') 
plt.ylabel('Total Sales') 
plt.title('Sales by Day of the Week') 
plt.xticks(rotation=45) 
plt.show()
```

## Visualization 16: Regarding scatterplots,  Linear Regression plots using Plotly.
