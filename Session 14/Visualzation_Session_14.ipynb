## Visualization 16: Regarding scatterplots,  Linear Regression plots using Plotly.
 We take the dips dataset, and we plot the linear relationship between total bills and tips.

```python
#linear regression
df = px.data.tips()   # Download data from plotly
df
```

```python
fig = px.scatter(df,
                 x="total_bill", y="tip",
                 trendline="ols",
                 width =600,height = 600
                 
                 )  # Ordinary least square
fig.show()
```
