#### Slicing

In addition to accessing list elements one at a time, Python provides concise syntax to access sublists; this is known as slicing:

```python
range(0,8,1)  # Start  End exclusive   Step 
```

```python
list(range(2,10,3)) # st 2 end 10  step 3 
```

```python
list(range(2,6,2))  # Start End step 
```

End is exclusive 

```python
nums = list(range(5,0,-1))  #Start 0  step 1   # range is a built-in function that creates a list of integers
print(nums)         # Prints 
```

```python
print(nums[2:4]) # : colon  To      Last index is exclusive
```

```python

# print(nums[2:4]) # Get a slice from index 2 to 4 (exclusive); prints 
# print(nums[2:])      # Get a slice from index 2 to the end; prints 
# print(nums[:2])     # Get a slice from the start to index 2 (exclusive); prints 
# print(nums[:])             # Get a slice of the whole list; prints 
# print(nums[:-1])    # Slice indices can be negative; prints 
print(nums)
nums[2:4] = [8, 9]                          # Assign a new sublist to a slice
print(nums)         # Prints "[0, 1, 8, 9, 4]"
```
